# What is Gradient Descent?

Gradient Descent is an optimization algorithm used to minimize a function (often a loss function in machine learning). It's an iterative method that moves in the direction of steepest descent as defined by the negative of the gradient.

Here's a basic outline of how Gradient Descent works:

1. Initialize the parameters of the model randomly.
2. Calculate the gradient of the loss function with respect to the parameters. The gradient is a vector that points in the direction of steepest ascent.
3. Update the parameters by taking a step in the direction of the negative gradient (i.e., the direction of steepest descent). The size of the step is determined by the learning rate, a hyperparameter that you must set.
4. Repeat steps 2 and 3 until the algorithm converges to a minimum.

The formula for the parameter update is:

```python
θ = θ - α * ∇J(θ)
```



Where:

- `θ` represents the parameters.
- `α` is the learning rate.
- `∇J(θ)` is the gradient of the loss function `J` with respect to the parameters `θ`.

There are also variants of Gradient Descent:

- **Batch Gradient Descent**: Uses the entire training set to compute the gradient at each step.
- **Stochastic Gradient Descent (SGD)**: Uses only a single training example to compute the gradient at each step.
- **Mini-batch Gradient Descent**: A compromise between Batch Gradient Descent and SGD. It uses a mini-batch of `n` training examples at each step.

Choosing the right variant and setting the learning rate properly is crucial for the algorithm to converge quickly and accurately.

# How Does Gradient Descent Work?

```python
Image(filename='images/GD.png')
```

Gradient Descent works by iteratively adjusting the parameters of a function to minimize the output of the function. Here's a step-by-step explanation:

1. **Initialization**: Start with random values for the parameters.

2. **Compute Gradient**: Calculate the gradient of the cost function at the current point. The gradient is a vector that points in the direction of the steepest ascent—i.e., the direction that increases the function most quickly. To minimize the function, we want to go in the opposite direction, so we will move in the direction of the negative gradient.

3. **Update Parameters**: Adjust the parameters by a small step in the direction of the negative gradient. The size of the step is determined by the learning rate, a hyperparameter that you must set.

4. **Iterate**: Repeat steps 2 and 3 until the algorithm converges to a minimum. Convergence is often determined by the gradient becoming very small (the function is flat) or by reaching a maximum number of iterations.

In the context of machine learning, the function we want to minimize is the cost function, which measures the error of our model. The parameters are the weights and biases in the model. By using Gradient Descent, we can iteratively adjust the weights and biases to find the best values that minimize the error of our model on the training data.

## Math Example

Let's consider a simple example where we have a cost function `J(θ) = θ^2`. We want to find the value of `θ` that minimizes `J(θ)`. We know from calculus that the minimum is at `θ = 0`, but let's use Gradient Descent to find this.

The gradient (derivative) of `J(θ)` is `2θ`. The update rule for Gradient Descent is `θ = θ - α * ∇J(θ)`, where `α` is the learning rate and `∇J(θ)` is the gradient of `J` at `θ`.

Let's start with `θ = 4` and `α = 0.1`. Here's how the first few iterations of Gradient Descent would look:

1. Calculate gradient: `∇J(θ) = 2θ = 2*4 = 8`.
2. Update `θ`: `θ = θ - α * ∇J(θ) = 4 - 0.1 * 8 = 3.2`.
3. Calculate gradient: `∇J(θ) = 2θ = 2*3.2 = 6.4`.
4. Update `θ`: `θ = θ - α * ∇J(θ) = 3.2 - 0.1 * 6.4 = 2.56`.

And so on. If you continue this process, you'll see that `θ` gets closer and closer to 0, which is the minimum of `J(θ)`. This is a simple example, but it illustrates the basic process of Gradient Descent. In practice, the cost function can be much more complex and can have many more parameters, but the basic idea is the same.
